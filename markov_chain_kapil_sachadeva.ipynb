```python
import numpy as np

# 초기 분포(Initial Distribution)
pi_0 = np.array([0.25, 0.25, 0.25, 0.25])

# 전이(transition) 확률 행렬
P = np.array([
    [0.00, 0.90, 0.10, 0.00],
    [0.80, 0.10, 0.00, 0.10],
    [0.00, 0.50, 0.30, 0.20],
    [0.10, 0.00, 0.00, 0.90]
])

def calculate_state_distribution(pi_0, P, N):
    pi_N = np.copy(pi_0)
    for _ in range(N):
        pi_N = np.dot(pi_N, P)#스칼라 곱
    return pi_N

N = int(input("N 값 넣기: "))

# N번의 전이(transition) 한 뒤의 분포 계산
result_distribution = calculate_state_distribution(pi_0, P, N)
print(f"State distribution after {N} transitions:\n{result_distribution}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 초기 분포(Initial Distribution)
pi_0 = np.array([0.25, 0.25, 0.25, 0.25])

# 전이(transition) 확률 행렬
P = np.array([
    [0.00, 0.90, 0.10, 0.00],
    [0.80, 0.10, 0.00, 0.10],
    [0.00, 0.50, 0.30, 0.20],
    [0.10, 0.00, 0.00, 0.90]
])

def calculate_state_distribution(pi_0, P, N):
    pi_N = np.copy(pi_0)
    for _ in range(N):
        pi_N = np.dot(pi_N, P)#스칼라 곱
    return pi_N

N = int(input("N 값 넣기: "))

# N번의 전이(transition) 후의 분포 계산
result_distribution = calculate_state_distribution(pi_0, P, N)

# 그래프로 보여주기
states = ['s1', 's2', 's3', 's4']
plt.bar(states, result_distribution)
plt.xlabel('States')
plt.ylabel('Probability')
plt.title(f'State Distribution after {N} Transitions')
plt.show()
```
