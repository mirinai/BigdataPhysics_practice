```python
#practice-4
'''
1. Write a function for 1-D random walk
1) An agent is at ùë• = 0 at ùë° = 0
2) with probability 0.5, agent moves forward(+1), otherwise, agent moves backward(-1)
t increase by 1
hint: use np.random.choice([-1,1]) or np.random.random()<0.5
3) repeat 2 until ùë°< ùë°ùëöùëéùë•
4) your input argument is ùë°ùëöùëéùë•, and your outcome is x_arr =
ùëõùëù. ùëéùëüùëüùëéùë¶( ùë• 0 , ùë• 1 , ùë• 2 , ‚Ä¶ , ùë• ùë°ùëöùëéùë• ). write a function for 1-D random walk
'''
import numpy as np
def OneD_random_walk(max):
  x_arr=np.zeros(max)
  for i in range(max):
    if i+1==max:
      break
    x_arr[i+1]=x_arr[i]+np.random.choice([-1,1],p=[0.5,0.5])
  return x_arr
max=100
print(OneD_random_walk(max))
print("-----------------------")
'''
2. Obtain < ùë•
2 > in 1-D random walk
1) Create empty 2-D array with shape (n, tmax)
hint: use np.zeros()
2) for i in range(n), calculate (x_arr)^2 and put the result to the 2-D array
hint: arr[i]=x_arr**2
3) Now, calculate average of ùë•^2
.
hint: arr.mean(axis=0)
.
'''
def TwoDArrayWithShape(n, tmax):
    arr = np.zeros((n, tmax)) # Create empty 2-D array
    for i in range(n):
        x_arr = OneD_random_walk(tmax)
        arr[i] = x_arr ** 2 # Calculate (x_arr)^2 and put the result to the 2-D array
    mean_x = arr.mean(axis=0) # Calculate the mean of x^2 for each time step
    return mean_x

n = 10
tmax = 50
result=TwoDArrayWithShape(n, tmax)
for i in range(tmax):
  if i%5==0:
    print()
  print("%5.1f"%(result[i]),end=" ")


```
