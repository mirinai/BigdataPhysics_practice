```python
import requests
import json
import pandas as pd

url="https://smart.incheon.go.kr/server/rest/services/Hosted/군구별_고령인구비율_정보/FeatureServer/0/query?where=1%3D1&outFields=*&outSR=4326&f=json"
res = requests.get(url)
r=json.loads(res.text)
print(r.keys())
print("\n")

for i in r.keys():
  print(i)
```

```python
#r["features"]
a_list,b_list=[],[]
for i in range(10):
  attr=r['features'][i]['attributes']
  a,b = attr['f2017'],attr['sig_kor_nm']
  a_list.append(a)
  b_list.append(b)
older=pd.DataFrame({'name':b_list,'2017':a_list})
older
```

```python
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen

target_url="https://sites.google.com/view/compsysbdai/publicatations"
res=urlopen(target_url)
html = res.read()

soup=bs(html, "html.parser")
print(soup)
```

```python
table = soup.find_all('span',{'class':"C9DxTc"})
print(table)
```

```python
for t in table:
  print(t)
```

```python
import re
for t in table:
  a=str(t)
  b=re.split(";|\xa0|>|<|</span|,",a)
  print(b)
```

```python
#save strings in list
cr=[]
for t in table:
  a=str(t)
  b=re.split(";|\xa0|>|<|</span|,",a)
  cr=cr+b[2:-2]
print(cr)
```
